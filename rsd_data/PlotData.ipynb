```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
plt.rcParams.update({'font.size': 12})
```

```python
# First plot the power spectra
cov3 = np.loadtxt('covariances/cov_joint_NGCSGCXi_z3.dat')
sigma3 = np.diag(cov3)**0.5

pk_ngc3 = np.loadtxt('pk/pk_NGC_z3.dat')
pk_sgc3 = np.loadtxt('pk/pk_SGC_z3.dat')

xi_3 = np.loadtxt('xi/z3.xi')
```

```python
cov1 = np.loadtxt('covariances/cov_joint_NGCSGCXi_z1.dat')
sigma1 = np.diag(cov1)**0.5

pk_ngc1 = np.loadtxt('pk/pk_NGC_z1.dat')
pk_sgc1 = np.loadtxt('pk/pk_SGC_z1.dat')

xi_1 = np.loadtxt('xi/z1.xi')
```

```python
# Prereconstruction xis
xip_1 = np.loadtxt('prerecon/BOSS_DR12_z038_s01.xis')
covp_1 = np.loadtxt('prerecon/BOSS_DR12_z038_s01.xiscov')
sigmap1 = np.diag(covp_1)**0.5
```

```python
xip_3 = np.loadtxt('prerecon/BOSS_DR12_z061_s01.xis')
covp_3 = np.loadtxt('prerecon/BOSS_DR12_z061_s01.xiscov')
sigmap3 = np.diag(covp_3)**0.5
```

```python
plt.figure(figsize=(10,7))

plt.subplot(2,2,1)

plt.errorbar(pk_ngc1[:,0], pk_ngc1[:,0] * pk_ngc1[:,1],\
             yerr=pk_ngc1[:,0] * sigma1[:40], fmt='o', c='C0')

plt.errorbar(pk_ngc1[:,0], pk_ngc1[:,0] * pk_ngc1[:,2],\
             yerr=pk_ngc1[:,0] * sigma1[40:80], fmt='o', c='C1')

plt.errorbar(pk_sgc1[:,0], pk_sgc1[:,0] * pk_sgc1[:,1],\
             yerr=pk_sgc1[:,0] * sigma1[80:120], fmt='o', fillstyle='none', c='C0')

plt.errorbar(pk_sgc1[:,0], pk_sgc1[:,0] * pk_sgc1[:,2],\
             yerr=pk_sgc1[:,0] * sigma1[120:160], fmt='o', fillstyle='none', c='C1')

plt.plot(-1000,1000,'C0',label=r'$\ell=0$')
plt.plot(-1000,1000,'C1',label=r'$\ell=2$')
plt.text(0.15, 50, 'z = 0.38')

plt.xlim(0,0.2)
plt.ylim(-100,2300)

plt.xticks([0,0.05,0.10,0.15,0.20])
plt.ylabel(r'Pre-Recon: $kP_\ell(k)$ [(Mpc/$h$)$^2$')
plt.xlabel(r'$k$ [$h^{-1}$ Mpc]' )

plt.legend()

plt.subplot(2,2,2)


plt.errorbar(pk_ngc3[:,0], pk_ngc3[:,0] * pk_ngc3[:,1],\
             yerr=pk_ngc3[:,0] * sigma3[:40], fmt='o', c='C0')

plt.errorbar(pk_ngc3[:,0], pk_ngc3[:,0] * pk_ngc3[:,2],\
             yerr=pk_ngc3[:,0] * sigma3[40:80], fmt='o', c='C1')

plt.errorbar(pk_sgc3[:,0], pk_sgc3[:,0] * pk_sgc3[:,1],\
             yerr=pk_sgc3[:,0] * sigma3[80:120], fmt='o', fillstyle='none', c='C0')

plt.errorbar(pk_sgc3[:,0], pk_sgc3[:,0] * pk_sgc3[:,2],\
             yerr=pk_sgc3[:,0] * sigma3[120:160], fmt='o', fillstyle='none', c='C1')

plt.plot(-1000,1000,'ko',label='NGC')
plt.plot(-1000,1000,'ko',fillstyle='none',label='SGC')
plt.text(0.15, 50, 'z = 0.61')

plt.xlim(0,0.2)
plt.ylim(-100,2300)
plt.xticks([0,0.05,0.10,0.15,0.20])
plt.yticks([])
plt.xlabel(r'$k$ [$h^{-1}$ Mpc]' )

plt.legend()

plt.subplot(2,2,3)

plt.errorbar(xi_1[:,0], xi_1[:,0]**2 * xi_1[:,1], yerr=xi_1[:,0]**2*sigma1[160:196], fmt='d', c='C0')
plt.errorbar(xi_1[:,0], -xi_1[:,0]**2 * xi_1[:,2], yerr=xi_1[:,0]**2*sigma1[196:232], fmt='d', c='C1')

plt.errorbar(xip_1[:,0], xip_1[:,0]**2 * xip_1[:,1], yerr=xip_1[:,0]**2*sigmap1[:24], fmt='x', c='C0')
plt.errorbar(xip_1[:,0], -xip_1[:,0]**2 * xip_1[:,2], yerr=xip_1[:,0]**2*sigmap1[24:], fmt='x', c='C1')

plt.text(120, -50, 'z = 0.38')
plt.xlabel(r'r [Mpc/$h$]')
plt.ylabel(r'$i^\ell r^2 \xi_\ell$ [(Mpc/$h$)$^2$')


plt.xlim(20,150)
plt.ylim(-60,140)


plt.subplot(2,2,4)

plt.errorbar(xi_3[:,0], xi_3[:,0]**2 * xi_3[:,1], yerr=xi_3[:,0]**2*sigma3[160:196], fmt='d', c='C0')
plt.errorbar(xi_3[:,0], -xi_3[:,0]**2 * xi_3[:,2], yerr=xi_3[:,0]**2*sigma3[196:232], fmt='d', c='C1')

plt.errorbar(xip_3[:,0], xip_3[:,0]**2 * xip_3[:,1], yerr=xip_3[:,0]**2*sigmap3[:24], fmt='x', c='C0')
plt.errorbar(xip_3[:,0], -xip_3[:,0]**2 * xip_3[:,2], yerr=xip_3[:,0]**2*sigmap3[24:], fmt='x', c='C1')

plt.yticks([])
plt.xlim(20,150)
plt.ylim(-60,140)

plt.plot(-1000,1000,'kd',label='Rec')
plt.plot(-1000,1000,'kx',label='Raw')
plt.legend()

plt.text(120, -50, 'z = 0.61')
plt.xlabel(r'r [Mpc/$h$]')

plt.tight_layout(h_pad = 0.1, w_pad=0.05)
plt.savefig('data_vector.pdf')
```

```python
# Now Plot Covariance Matrices
krange = (pk_ngc1[:,0] > 0.02) * (pk_ngc1[:,0] < 0.2)
rrange = (xi_1[:,0] > 80) * (xi_1[:,1] < 130)
iis = np.concatenate( (krange, krange, krange, krange, rrange, rrange) )
```

```python
plt.figure(figsize=(10,4.5))

plt.subplot(1,2,1)

cov_fit = cov1[np.ix_(iis,iis)]
sigmas = np.sqrt(np.diag(cov_fit))
corr_fit = cov_fit / sigmas[:,None] / sigmas[None,:]

plt.imshow(corr_fit, cmap='coolwarm', vmin=-0.5, vmax=1.0)

plt.yticks([])
plt.ylabel(r'$\quad \xi_2 \qquad  \xi_0 \qquad P^{\rm SGC}_2 \quad \; P^{\rm SGC}_0 \quad \; P^{\rm NGC}_2 \quad \; P^{\rm NGC}_0$')

plt.title('z = 0.38')

plt.subplot(1,2,2)

cov_fit = cov3[np.ix_(iis,iis)]
sigmas = np.sqrt(np.diag(cov_fit))
corr_fit = cov_fit / sigmas[:,None] / sigmas[None,:]

plt.imshow(corr_fit, cmap='coolwarm', vmin=-0.5, vmax=1.0)
plt.colorbar()

plt.title('z = 0.61')
plt.yticks([])
#plt.ylabel(r'$\quad \; \xi_2 \qquad  \xi_0 \qquad P^{\rm SGC}_2 \quad \; P^{\rm SGC}_0 \quad \; P^{\rm NGC}_2 \quad \; P^{\rm NGC}_0$')

plt.tight_layout()
plt.savefig('covariance.pdf')
```
